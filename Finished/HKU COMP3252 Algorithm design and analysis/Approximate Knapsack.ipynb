# 1. Definition

We define:
+ **Knapsack problem**:
  + Input: A set of items, each with a weight $w_i$ and a value $v_i$, and a maximum weight capacity $W$.
  + Output: A subset of items that maximizes the total value while not exceeding the weight capacity.
  + **OPT**: 
    + The maximum total value that can be achieved with the given constraints.
    + i.e. $OPT = \max \left\{ \sum_{i\in S}^n v_i \mid S\subseteq \{1,2,\ldots,n\}, \sum_{i\in S} w_i \leq W \right\}$
Then we define:
+ **C-Approximation algorithm**:
  + An algorithm that finds a solution within a factor of $C$ of the optimal solution.
  + i.e. For C-Approximation Knapsack algorithm, it finds a solution with value that $\ge \frac{OPT}{C}$.

# 2. Algorithms for approximate Knapsack problem

## 2.1 Greedy Algorithm
We consider the following greedy algorithm for the knapsack problem:
1. Sort the items by their value-to-weight ratio $r_i = \frac{v_i}{w_i}$ in descending order.
2. Initialize an empty set $S$ and a variable $total\_weight = 0$.
3. Iterate through the sorted items:
   - For each item $i$, if adding it to the set $S$ does not exceed the weight capacity $W$, add it to $S$ and update $total\_weight$.
   - Otherwise, skip the item and move to the next one.

So what's the approximation ratio of this greedy algorithm? Consider the following cases:
> There are two items, each item only can be taken once.
> - Item 1: $v = 2\epsilon$, $w = \epsilon$
> - Item 2: $v = 1$, $w = 1$
> And the weight capacity $W = 1$.    
>
> The greedy algorithm will select item 1 and get a total value of $2\epsilon$. 
> So the C is $\frac{1}{2\epsilon}$, which can be arbitrarily large as $\epsilon$ approaches 0.

Therefore, the approximation ratio of the greedy algorithm can be arbitrarily bad, and it is not a C-Approximation algorithm for any constant $C$.

## 2.2 Improved Greedy Algorithm

We still sort the items by their value-to-weight ratio $r_i = \frac{v_i}{w_i}$ in descending order. But instead of just taking items until we can't, we also consider the single item with the highest value. The algorithm is as follows:
1. Sort the items by their value-to-weight ratio $r_i = \frac{v_i}{w_i}$ in descending order.
2. Greedily select items in this order until we cannot add any more items without exceeding the weight capacity $W$. We donate this item's index as $k$.
   + i.e. $\sum^{k-1}_{i=1} w_i \leq W$
3. We return the better of the two solutions: 
   1. The greedy solution (items $1$ to $k-1$)
   2. The single item with the highest value (item $1$).

By this improved greedy algorithm, we can guarantee that the approximation ratio is at most 2. This is because:
Proof:

We consider the "fractional knapsack problem", where we can take fractions of items. 
> i.e. We can take a fraction $p_i$ of item $i$: $p_i \in [0, 1]$

Then we donate the optimal solution of the fractional knapsack problem as $OPT_f$. We can see that $OPT_f \geq OPT$, because the fractional knapsack problem is a relaxation of the original knapsack problem.

Now we consider the greedy solution of the fractional knapsack problem, obviously is that:
$$
OPT_f = \sum^{k-1}_{i=1} v_i + v_k \cdot \frac{W - \sum^{k-1}_{i=1} w_i}{w_k} \leq \sum^{k-1}_{i=1} v_i + v_k
$$

Therefore, we have:
$$
\text{Improved Greedy Solution} = \max \left\{ \sum^{k-1}_{i=1} v_i, v_1 \right\} \geq \frac{OPT_f}{2} \geq \frac{OPT}{2}
$$

Thus the improved greedy algorithm is a 2-Approximation algorithm for the knapsack problem.



## 2.3 Fully Polynomial-Time Approximation Scheme (FPTAS)

The idea of FPTAS is to scale down the values of the items to reduce the problem size, and then apply a dynamic programming algorithm to solve the scaled problem. The steps are as follows:
1. Let $V = \max_{i} v_i$ be the maximum value of the items.
2. Choose a scaling factor 
   + $$K = \frac{\epsilon V}{n}$$
   + Where $\epsilon$ is the desired approximation ratio (e.g., $\epsilon = 0.1$ for a 10% approximation).
3. Scale down the values of the items: $$\tilde{v}_i = \lfloor \frac{v_i}{K} \rfloor$$ for each item $i$.
4. Use a dynamic programming algorithm to solve the knapsack problem with the scaled values $\tilde{v}_i$ and the original weights $w_i$.
5. The solution obtained from the dynamic programming algorithm will be an approximate solution to the original knapsack problem.

Because we have scaled down the values, the dynamic programming algorithm will run in polynomial time: $Poly(n, \frac{1}{\epsilon}) = O(\frac{n^3}{\epsilon})$. Now we consider the approximation ratio of this algorithm:

Beacuse the values has been scaled down, each item's value lost at most $K$. Therefore, the total value lost by the approximation is at most $nK$. So we have:
$$
\text{Approximate Solution} \geq OPT - nK = OPT - n \cdot \frac{\epsilon V}{n} = (1-\epsilon)OPT
$$
The $\epsilon$ always small enough, so that $(1-\epsilon)^{-1}\approx 1 + \epsilon$. Therefore we can say that the approximation ratio is at most $(1 + \epsilon)$. Thus the FPTAS is a $(1 + \epsilon)$-Approximation algorithm for the knapsack problem.
