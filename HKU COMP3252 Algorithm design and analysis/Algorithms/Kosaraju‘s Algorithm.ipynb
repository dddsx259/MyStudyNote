# Kosaraju's algorithm

## 1. We have the following definitions:
+ **Connected Components, CC**
  + For undirected graphs
  + A connected component is a maximal set of vertices such that there is a path between every pair of vertices in the set. i.e. Each pair of vertices in the set is connected.
+ **Strongly Connected Components, SCC**
  + For directed graphs
  + A strongly connected component is a maximal set of vertices such that there is a path from every vertex in the set to every other vertex in the set. i.e. Each pair of vertices in the set is **strongly connected**.
  + Strongly connected means that there is a path from vertex A to vertex B and also a path from vertex B to vertex A.
+ **Meta graph**
  + Consider the connectness, all vertics in the same SCC can be **treated as a single vertex** in a manner.
  + Note that each vertex in the original graph belongs to exactly one SCC, which means that the graph can be partitioned into SCCs. 
  + The meta graph is a directed acyclic graph (DAG) where each vertex represents an SCC from the original graph, and there is a directed edge from vertex A to vertex B in the meta graph if there is a directed edge from any vertex in SCC A to any vertex in SCC B in the original graph.
  

## 2. Kosaraju's algorithm
So we want to efficiently find all the SCCs in a directed graph. Kosaraju's algorithm is a two-pass depth-first search (DFS) algorithm that can find all the SCCs in linear time, $O(V + E)$, where $V$ is the number of vertices and $E$ is the number of edges in the graph.

1. We perform a DFS on the original graph. Each time we finish processing a vertex (i.e., when we backtrack from it), we record its finishing time, and we push it onto a stack. 
2. We reverse the graph, which means we reverse the direction of all edges.
   1. crucial character: The SCCs in the original graph are the same as the SCCs in the reversed graph. 
3. Repeatly pop the top element in the stack:
   1. If the vertices isn't visited in the reversed graph $G^r$:
      1. We perform a DFS starting from that vertex in the reversed graph. All the vertices reachable from that vertex in the reversed graph form a strongly connected component. We can mark all these vertices as visited to avoid processing them again.
      2. We repeat this process until the stack is empty. Each time we perform a DFS in the reversed graph, we will find one strongly connected component.

```python
from myGraph import Graph

example_graph = Graph(10, graph_type='directed')
edges = [
    [1,2], [1,7], [2,3], [3,5], [3,8], [3,9], [4,3], [5,6], [6,9], [7,8], [8,1], [9,5]
]
example_graph.add_edges(edges)

def Kosaraju(graph):
    visited = set()
    stack = []
    sccs = []

    def dfs(v):
        visited.add(v)
        for neighbor in graph.get_neighbors(v):
            if neighbor not in visited:
                dfs(neighbor)
        stack.append(v)

    def reverse_dfs(v, current_scc):
        visited.add(v)
        current_scc.append(v)
        for neighbor in reversed_graph.get_neighbors(v):
            if neighbor not in visited:
                reverse_dfs(neighbor, current_scc)

    # Step 1: Perform DFS and fill the stack
    for vertex in graph.get_vertices():
        if vertex not in visited:
            dfs(vertex)

    # Step 2: Reverse the graph
    edges = graph.get_all_edges()
    reversed_graph = Graph(graph.n_vertices, graph_type='directed')
    reversed_graph.add_edges([(v, u) if not graph.weighted else (v, u, w) for u, v, *w in edges])

    # Step 3: Perform DFS on the reversed graph using the order in the stack
    visited.clear()
    while stack:
        vertex = stack.pop()
        if vertex not in visited:
            current_scc = []
            reverse_dfs(vertex, current_scc)
            sccs.append(current_scc)

    return sccs


print("Strongly Connected Components:", Kosaraju(example_graph))
```
