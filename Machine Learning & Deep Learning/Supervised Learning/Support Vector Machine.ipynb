# Support Vector Machine(SVM)


## 1. Margin and Support Vectors
The **margin** of a hyperplane is defined as the distance from the hyperplane to the nearest data point. The **support vectors** are the data points that are closest to the hyperplane and lie on the margin.(as shown in the figure below)



The core idea of SVM is to find the hyperplane that maximizes the margin, which is known as the **maximum margin hyperplane**. By maximizing the margin, SVM aims to improve the generalization ability of the model and reduce the risk of overfitting.    
The width of the "street":
$$
(\mathbf{x}_+-\mathbf{x}_-)^T\frac{\mathbf{w}}{\|\mathbf{w}\|} = \frac{2}{\|\mathbf{w}\|}
$$
i.e. we want to maximize the width of the "street"
$$
\max_{\mathbf{w}, b} \frac{2}{\|\mathbf{w}\|}
$$
which is equivalent to minimizing $\|\mathbf{w}\|$. Therefore, the optimization problem can be formulated as:
$$
\min_{\mathbf{w}, b} \frac{1}{2}\|\mathbf{w}\|^2
$$
subject to $y_i(\mathbf{w} \cdot \mathbf{x_i} + b) \geq 1$ for all $i$(If the data is **linear separable**). This optimization problem can be solved using techniques such as quadratic programming, and the resulting hyperplane is determined by the support vectors.

## 2. Optimization Problem
The optimization problem for SVM can be expressed as:
$$
\min_{\mathbf{w}, b} \frac{1}{2}\|\mathbf{w}\|^2
$$
subject to $y_i(\mathbf{w} \cdot \mathbf{x_i} + b) \geq 1$ for all $i$. This is a convex optimization problem that can be solved using various optimization algorithms, such as gradient descent or quadratic programming. The solution to this optimization problem will yield the optimal hyperplane that maximizes the margin between the two classes in the feature space.      

## 3. Non-linear SVM and Kernel Trick
In cases where the data is not linearly separable, SVM can be extended to handle non-linear decision boundaries using the **kernel trick**. The kernel trick allows SVM to implicitly map the input data into a higher-dimensional feature space where a linear separation is possible. This is achieved by defining a kernel function that computes the inner product of the data points in the feature space without explicitly performing the mapping. Common kernel functions include the polynomial kernel, radial basis function (RBF) kernel, and sigmoid kernel. By using the kernel trick, SVM can effectively capture complex relationships between the datapoints and improve classification performance in non-linear scenarios.

### Usual kernels:
- **Linear kernel**: $K(\mathbf{x}, \mathbf{y}) = \mathbf{x}^T \mathbf{y}$
- **Polynomial kernel**: $K(\mathbf{x}, \mathbf{y}) = (\mathbf{x}^T \mathbf{y} + c)^d$
- **Radial basis function (RBF) kernel**: $K(\mathbf{x}, \mathbf{y}) = \exp(-\gamma \|\mathbf{x} - \mathbf{y}\|^2)$
- **Sigmoid kernel**: $K(\mathbf{x}, \mathbf{y}) = \tanh(\alpha \mathbf{x}^T \mathbf{y} + c)$

# 4. Hard Margin vs Soft Margin SVM
Some time the data is not perfectly linearly separable, and in such cases, we can use a **soft margin SVM**. The soft margin SVM allows for some misclassification of the training data by introducing slack variables $\xi_i$ that measure the degree of misclassification for each data point. The optimization problem for soft margin SVM can be formulated as:
$$\min_{\mathbf{w}, b, \xi} \frac{1}{2}\|\mathbf{w}\|^2 + C \sum_{i=1}^{N} \xi_i$$
subject to $y_i(\mathbf{w} \cdot \mathbf{x_i} + b) \geq 1 - \xi_i$ for all $i$, and $\xi_i \geq 0$ for all $i$.    

Here, $C$ is a **regularization parameter** that controls the trade-off between maximizing the margin and minimizing the classification error. 
+ A larger value of $C$ will result in a smaller margin and fewer misclassifications
+ Asmaller value of $C$ will allow for a larger margin but may result in more misclassifications.    

The choice between hard margin and soft margin SVM depends on the specific characteristics of the dataset and the desired balance between margin maximization and classification accuracy.
