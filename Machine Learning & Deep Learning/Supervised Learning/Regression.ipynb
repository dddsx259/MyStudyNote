# Regression

Regression is a type of supervised learning algorithm that is used to predict a continuous output variable based on one or more input variables. The goal of regression is to find the best-fitting line or curve that describes the relationship between the input variables and the output variable.

## 1. Linear Regression

### 1.1 Simple Linear Regression
+ Simple linear regression is a type of regression that models the relationship between a single input variable (independent variable) and a single output variable (dependent variable) using a straight line. The equation of the line is given by:
+ $y = wx + b$, where:
  + $y$ is the predicted output
  + $x$ is the input variable
  + $w$ is the weight (slope)
  + $b$ is the bias (intercept).
+ The goal is to find the weight and bias that minimize the mean squared error between the predicted values and the actual values.
+ Geometrically, we want to find the line that best fits the data points in the input-output space. 

### 1.2 Multiple Linear Regression
+ Multiple linear regression is an extension of simple linear regression that models the relationship between multiple input variables and a single output variable. The equation for multiple linear regression is given by:
+ $y = \vec{w}\vec{x} + b$, where:
  + $y$ is the predicted output
  + $\vec{x}$ is the vector of input variables
  + $\vec{w}$ is the vector of weights (coefficients)
  + $b$ is the bias (intercept).
+ The goal is to find the weights and bias that minimize the mean squared error between the predicted values and the actual values.
+ Geometrically, we want to find the hyperplane that best fits the data points in the multi-dimensional input-output space. 

## 2. Exploratory Data Analysis (EDA)
+ Before applying regression models, it is important to perform exploratory data analysis (EDA) to understand the data and identify any patterns or relationships between the input variables and the output variable.

### 2.1 Scatterplots Matrix
+ A scatterplot matrix is a grid of scatterplots that shows the pairwise relationships between multiple variables. 
+ We can use the scatterplotmatrix function from the MLxtend library

### 2.2 Correlation Matrix
+ A correlation matrix is a table that shows the **Correlation cofficiencire** between multiple variables.
+ The correlation coefficient ranges from -1 to 1, where:
  + A value of 1 indicates a perfect positive correlation (as one variable increases, the other variable also increases).
  + A value of -1 indicates a perfect negative correlation (as one variable increases, the other variable decreases).
  + A value of 0 indicates no correlation (there is no linear relationship between the variables).
  + $$\text{Correlation coefficient} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}$$
  + Where $\text{Cov}(X, Y)$ is the covariance between variables X and Y, and $\sigma_X$ and $\sigma_Y$ are the standard deviations of X and Y, respectively.
+ We can use the corr function from the pandas library to compute the correlation matrix.

#### Covariance
+ Covariance is a measure of how two variables change together. It is calculated as:
  + $$\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$
  + Where $X_i$ and $Y_i$ are the individual data points of variables X and Y, $\bar{X}$ and $\bar{Y}$ are the means of X and Y, and n is the number of data points.
+ If we only have a partial dataset, we can use the sample covariance, which is calculated as:
  + $$\text{Cov}(X, Y) = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$
  + The sample covariance is used to estimate the covariance of the entire population based on a sample of data.

## 3. Model training
+ After performing EDA, we can train a regression model on the data and evaluate its performance using appropriate metrics.

### 3.1 Oridinary Least Squares (OLS)
+ Ordinary Least Squares (OLS) is a method for estimating the parameters of a linear regression model. The OLS method minimizes **the sum of the squared** differences between the observed values and the predicted values.
+ Implements:
  1. Similar to Adaline, we can use gradient descent to find the optimal weights and bias that minimize the mean squared error.
  2. **Closed-form solution**:
     + We can derive a closed-form solution for the weights and bias by setting the gradient of the cost function to zero and solving for the parameters. 
     + Only works for linear regression and is computationally expensive for large datasets.

Note that OLS wants to fit all data, which means that it is sensitive to outliers. If there are outliers in the data, OLS may not perform well and may produce a model that does not generalize well to new data. 

### 3.2 Robust Regression
+ Robust regression is a type of regression that is designed to be less sensitive to outliers in the data. One popular method for robust regression is RANSAC (Random Sample Consensus).
+ The core idea of RANSAC is instead of fitting all the data, it randomly samples a subset of the data points and fits a model to this subset. It then determines how many data points are inliers (close to the fitted model) and how many are outliers (far from the fitted model). This process is repeated for a specified number of iterations, and the model with the highest number of inliers is selected as the final model.
+ Steps of RANSAC:
  1. Determine the minimum number of data points required to fit the model (e.g., for linear regression, we need at least 2 data points).
  2. Process of iteration:
     1. Randomly select the minimum number of data points from the dataset to fit the model.
     2. Fit the model to the selected data points.
     3. Calculate the distance of all data points to the fitted model, if the distance is less than a predefined threshold, count it as an inlier, otherwise count it as an outlier.
     4. Keep track of the model with the highest number of inliers.
  3. If find the model with the highest number of inliers or reach the maximum number of iterations, stop the process and return the best model.
  4. (Optional) Refit the model using all inliers to improve the accuracy of the final model.

## 4. Model Evaluation

+ After training a regression model, we need to evaluate its performance using appropriate metrics. Common metrics for evaluating regression models include:
  + **Residual Plot**: An ideal residual plot should show a random scatter of points around the horizontal axis (residuals = 0). If there is a pattern in the residual plot (e.g., a funnel shape), it may indicate that the model is not fitting the data well.
  + **Mean Absolute Error (MAE)**: The average absolute difference between the predicted values and the actual values.
    + $$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$
  + **Mean Squared Error (MSE)**: The average squared difference between the predicted values and the actual values.
    + $$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$
  + $R^2$ Score: The proportion of the variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, where a value of 1 indicates that the model explains all the variance in the data, and a value of 0 indicates that the model does not explain any of the variance.
    + $$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$
    + Where $y_i$ are the actual values, $\hat{y}_i$ are the predicted values, and $\bar{y}$ is the mean of the actual values.
    + $R^2 = 1$ indicates a perfect fit
    + $R^2 = 0$ indicates that the model does not explain any of the variance in the data
    + $R^2 < 0$ indicates that the model is worse than a horizontal line (i.e., the mean of the actual values).

## 5. Regularization
To avoid overfitting, we can use regularization techniques that add a **penalty term** to the cost function to discourage the model from fitting the noise in the data. Common regularization techniques include:
+ **Ridge Regression (L2 Regularization)**: Adds a penalty term that is proportional to the square of the magnitude of the coefficients. The cost function for ridge regression is given by:
  + $$J(w, b) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{m} w_j^2$$
  + Where $\lambda$ is the regularization parameter that controls the strength of the penalty.
+ **Lasso Regression (L1 Regularization)**: Adds a penalty term that is proportional to the absolute value of the coefficients. The cost function for lasso regression is given by:
  + $$J(w, b) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{m} |w_j|$$
  + Lasso regression can shrink some coefficients to zero, effectively performing feature selection.
+ **Elastic Net Regression**: Combines both L1 and L2 regularization. The cost function for elastic net regression is given by:
  + $$J(w, b) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda_1 \sum_{j=1}^{m} |w_j| + \lambda_2 \sum_{j=1}^{m} w_j^2$$
  + Where $\lambda_1$ and $\lambda_2$ are the regularization parameters that control the strength of the L1 and L2 penalties, respectively.

## 6. Non-linear Regression
+ Non-linear regression is a type of regression that models the relationship between the input variables and the output variable using a non-linear function. Common non-linear regression models include:
+ **Polynomial Regression**
  + Polynomial regression is a type of regression that models the relationship between the input variables and the output variable using a polynomial function. The equation for polynomial regression is given by:
    + $$y = w_0 + w_1 x + w_2 x^2 + ... + w_d x^d + b$$
    + Where $d$ is the degree of the polynomial, and $w_0, w_1, ..., w_d$ are the coefficients of the polynomial.
    + Essentially, polynomial regression is just a linear regression, just with additional features that are powers of the original input variable. 
+ Other non-linear functions like exponential, logarithmic, etc. can also be used for non-linear regression depending on the nature of the data and the relationship between the input and output variables.

## 7. Tree-based Regression
+ Tree-based regression models, such as decision trees and random forests, can also be used for regression tasks. These models can capture non-linear relationships between the input variables and the output variable without the need for feature engineering. Tree-based regression models work by recursively partitioning the input space into regions and fitting a simple model (e.g., a constant value) in each region. The final prediction is made by averaging the predictions from all the regions. Tree-based regression models can handle both numerical and categorical input variables and are less sensitive to outliers compared to linear regression models. However, they can be prone to overfitting if not properly regularized (e.g., by limiting the depth of the tree or using ensemble methods like random forests).

### 7.1 Decision Tree Regression
+ Decision tree regression model partitions the input space into regions and predicts a constant value for each region. 
+ The tree is built by recursively splitting the data based on the input variables to minimize the mean squared error in each region. 
+ The final prediction is made by averaging the predictions from all the regions.
+ Decision tree regression can capture non-linear relationships between the input variables and the output variable, but it can be prone to overfitting if the tree is too deep. 
+ To prevent overfitting, we can limit the depth of the tree or use ensemble methods like random forests, which combine multiple decision trees to improve the accuracy and robustness of the model.

+ The prediction value is the mean of all trees.
