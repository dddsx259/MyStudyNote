# 1. Introduction to Ensemble Learning

Ensemble learning is a powerful machine learning technique that combines the predictions of multiple models to improve overall performance. The **core idea** is that by **aggregating the outputs of several models**, we can reduce the likelihood of overfitting and increase the robustness of our predictions. As a chinese proverb says, "三个臭皮匠, 顶个诸葛亮".

There are three common types of ensemble methods:
+ Majority Voting
+ Bagging (Bootstrap Aggregating)
+ Boosting

# 2. Majority/Plurality & Weighted Voting

The core idea of majority voting is to take the predictions of multiple models and select the class that receives the most votes as the final prediction. 

## 2.1 Theoretical analysis

We suppose we have $n$ classifiers, each with an error rate of $\epsilon$, and we assume that the errors of the classifiers are independent. So if the accuracy of each classifier is bigger than randomly guessing, i.e., $\epsilon < 0.5$, then the ensemble of these classifiers will have a lower error rate than any individual classifier.

$$
P(\text{Ensemble makes an error}) = \sum_{k=\lceil \frac{n}{2} \rceil}^{n} \binom{n}{k} \epsilon^k (1-\epsilon)^{n-k} \rightarrow 0 \text{ as } n \rightarrow \infty
$$

## 2.2 Weighted Voting

Sometimes different classifiers may have different levels of accuracy, and we can assign weights to their predictions accordingly. In weighted voting, each classifier's vote is weighted by its accuracy or some other performance metric. The final prediction is then determined by the weighted sum of the votes.


# 3. Bagging(Bootstrap Aggregating)

**1. Core Concept**
Bagging is a powerful **Ensemble Learning** technique designed to improve the stability and accuracy of machine learning algorithms. It works by training multiple models in parallel and combining their predictions.

**2. How it Works (The Workflow)**
*   **Bootstrap Sampling:** It creates multiple subsets of the original training data by sampling **with replacement**. This means some data points may appear multiple times in a subset, while others are left out.
*   **Parallel Training:** A base learner (often a Decision Tree) is trained independently on each bootstrap sample.
*   **Aggregation:**
    *   **Classification:** Uses **Majority Voting** (the class with the most votes wins).
    *   **Regression:** Uses **Averaging** (the mean of all predictions).

**3. Key Characteristics**
*   **Reduces Variance:** Its primary goal is to lower the variance of a model, which effectively helps in preventing **overfitting**.
*   **Ideal Candidates:** It works best with "unstable" models that have high variance and low bias, such as deep Decision Trees.
*   **Out-of-Bag (OOB) Data:** Roughly 36.8% of the original data is not selected in a bootstrap sample. This "OOB data" can be used as a built-in validation set to test the model's performance.

**4. Classic Example**
*   **Random Forest:** The most famous application of Bagging. It builds upon the Bagging framework by also randomly selecting features at each split in the decision trees to further reduce correlation between models.



# 4. Boosting

The core idea of boosting is:
+ Train a sequence of weak learners.
  + Performance of each learner is a bit better than random guessing.
+ Each learner focuses on the mistake samples that the previous learners misclassified.
+ By doing so, the ensemble can learn from its mistakes and improve its performance over time.
+ The final prediction is a weighted combination of the predictions from all the learners.

## 4.1 The original boosting algorithm: 
1. Train $C_1$ on the original training data.
2. Train $C_2$ on the same training data, but with more weight on the samples that $C_1$ misclassified.
3. Train $C_3$ focus on the samples that $C_1$ and $C_2$ have different predictions.
4. Predict the final output by taking a majority vote of $C_1$, $C_2$, and $C_3$.

## 4.2 AdaBoost
1. Initialize the weights of the training samples to be equal.
   1. i.e. $\sum_{i=1}^{N} w_i = 1$.
2. For $t = 1$ to $T$:
   1. Train a weak learner $C_t$ on the training data with weights $w_i$.
   2. Calculate the predictions $\hat{y_t}$ of $C_t$ on the training data.
   3. Calculate the error rate of $C_t$: $\epsilon_t = \sum_{i=1}^{N} w_i \cdot I(C_t(x_i) \neq y_i)$, where $I$ is the indicator function.
   4. The vote weight of $C_t$ is calculated as $\alpha_t = \frac{1}{2} \ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right)$.
   5. Update the weights of the training samples:
      1. For correctly classified samples: $w_i \leftarrow w_i \cdot e^{-\alpha_t}$.
      2. For misclassified samples: $w_i \leftarrow w_i \cdot e^{\alpha_t}$.
   6. Normalize the weights so that they sum to 1: $w_i \leftarrow \frac{w_i}{\sum_{j=1}^{N} w_j}$.
3. The final prediction is made by taking a weighted majority vote of the weak learners: $\hat{y} = \text{sign}\left(\sum_{t=1}^{T} \alpha_t C_t(x)\right)$.
