# Logistic Regression

## 1. Definitions

### 1,1. logit/linear output z
$$
z = \vec{w}^T\vec{x} + b
$$

### 1.2. Sigmoid function
We map the linear output $z$ into section $(0,1)$, explain as probability:
$$
P(y=1\mid \vec{x}, \vec{w}) = \phi(z)=\frac{1}{1 + e^{-z}}
$$
So that:
$$
\begin{cases}
\phi(z)\rightarrow 1, &\text{when } z\rightarrow +\infty\\
\phi(z)\rightarrow 0, &\text{when } z\rightarrow -\infty
\end{cases}
$$

And we have:
+ If $P(y=1\mid \vec{x}, \vec{w}) > 0.5$, output $\hat{y}=1$
+ Otherwise, output $\hat{y}=0$

### 1.3. Cost function
We consider using **Likelihood** to describe the error:
For each sample $(\vec{x}^{(i)}, y^{(i)})$, we define:
$$
\begin{aligned}
L^{(i)}
&= L(y\mid \vec{x}, \vec{w})\\
&= (\phi(z)^{(i)})^{y^{(i)}}(1-\phi(z)^{(i)})^{1-y^{(i)}}\\
&=
\begin{cases}
\phi(z)^{(i)}, &if\ y=1\\
1-\phi(z)^{(i)}, &otherwise\\
\end{cases}
\end{aligned}
$$

To simplify the expression of total cost (instead of using product), we take the logarithm of the likelihood:
$$
\mathcal{L}^{(i)}= - \big[y^{(i)}\log(\phi(z)^{(i)}) + (1-y^{(i)})\log(1 - \phi(z)^{(i)})\big]\\
$$
The $\mathcal{L}$ we called **Log Loss/Cross-Entropy Loss**

Then we define the total cost:
### 1.4. Total cost
$$
J(\vec{w}, b ) = \frac{1}{m}\sum_{i=1}^m\mathcal{L}^{(i)}
$$
Which is a convex function, obviously we can use some optimization algorithm to update the weights.

## 2. Gradient Descent

Note that:
$$
\begin{aligned}
&\frac{\partial J}{\partial w_j} = \frac{1}{m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)}) x_j^{(i)}\\
&\frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})
\end{aligned}
$$

So we have:
$$
\begin{aligned}
&w_j := w_j - \alpha \frac{\partial J}{\partial w_j}\\
&b := b - \alpha \frac{\partial J}{\partial b}
\end{aligned}
$$
Where $\alpha$ is **Learning Rate**


### 3. Multi-class Logistic Regression

We use the **Softmax function** (also known as **Multinomial Logistic Regression**) to handle classification with $K \geq 2$ classes.

For each class $k \in \{1, 2, \dots, K\}$, we have a dedicated weight vector $\vec{w}_k$ and bias term $b_k$. For a given input sample $\vec{x}$, the logit (unnormalized log-probability) for class $k$ is:

$$
z_k = \vec{w}_k^T \vec{x} + b_k
$$

The model then outputs a probability distribution over the $K$ classes via the Softmax function:

$$
P(y = k \mid \vec{x}) = \frac{e^{z_k}}{\sum_{j=1}^{K} e^{z_j}}
$$

This ensures that:
- $P(y = k \mid \vec{x}) \geq 0$ for all $k$,
- $\sum_{k=1}^{K} P(y = k \mid \vec{x}) = 1$.

The corresponding loss function for a single sample is the **categorical cross-entropy**:
$$
\mathcal{L} = -\sum_{k=1}^{K} y_k \log\big(P(y = k \mid \vec{x})\big)
$$
where $y_k$ is the ground-truth label in one-hot encoding.
