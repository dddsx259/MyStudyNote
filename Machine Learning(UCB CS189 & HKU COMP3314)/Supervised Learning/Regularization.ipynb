# Step1. Standardization
$$
x_{std} = \frac{x-\mu}{\sigma}
$$

# Step2. Add Regularization Term to Loss function

$$
\text{Total Loss} = \text{Original Loss} + λ⋅\text{Regularization Term}
$$

There are three common regularization type:

## 2.1 L1 regularization(Lasso Regression)
$$
\text{Regularization Term} = ||\vec{w}||  = \sum w_i
$$


## 2.2 L2 regularization(Ridge Regression / Weight Decay)
$$
\text{Regularization Term} = ||\vec{w}||^2  = \sum w_i^2
$$

## 2.3 Elastic Net
$$
\text{Regularization Term} = \alpha||\vec{w}|| + (1-\alpha)\alpha||\vec{w}||^2  = \sum w_i^2
$$
